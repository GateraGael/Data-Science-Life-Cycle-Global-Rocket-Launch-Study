# Timeline of artificial satellites and space probes

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

artificialsats_spaceprobes_url = requests.get("https://en.wikipedia.org/wiki/Timeline_of_artificial_satellites_and_space_probes").text

soup = BeautifulSoup(artificialsats_spaceprobes_url,"lxml")
my_tables = soup.find_all("table",{"class":"wikitable"})
```

```python
satellites_probes_1950_df = pd.DataFrame(pd.read_html(str(my_tables[0]))[0])
print(satellites_probes_1950_df.columns)
satellites_probes_1960_df = pd.DataFrame(pd.read_html(str(my_tables[1]))[0])
print(satellites_probes_1960_df.columns)
```

```python
columns_list = list(pd.DataFrame(pd.read_html(str(my_tables[2]))[0]).columns)
satellites_probes_df = pd.DataFrame(columns=columns_list)

for table in my_tables[2:]:
    # read the HTML as a string
    html_list = pd.read_html(str(table))

    # convert list to dataframe
    df = pd.DataFrame(html_list[0])
    
    # concactenate original dataframe with tables 
    satellites_probes_df = pd.concat([satellites_probes_df, df], ignore_index=True)

satellites_probes_df.shape
```

# Rocket Launch Sites Wiki

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

artificialsats_spaceprobes_url = requests.get("https://en.wikipedia.org/wiki/List_of_rocket_launch_sites#cite_note-88").text

soup = BeautifulSoup(artificialsats_spaceprobes_url,"lxml")
all_launch_sites_tables = soup.find_all("table",{"class":"wikitable"})
```

```python
launch_sites_columns_list = list(pd.DataFrame(pd.read_html(str(all_launch_sites_tables[0]))[0]).columns)
launch_sites_df = pd.DataFrame(columns=launch_sites_columns_list)

for table in all_launch_sites_tables:
    # read the HTML as a string
    html_list = pd.read_html(str(table))

    # convert list to dataframe
    df = pd.DataFrame(html_list[0])
    
    # concactenate original dataframe with tables 
    launch_sites_df = pd.concat([launch_sites_df, df], ignore_index=True)
```

```python
launch_sites_df.loc[0,'Coordinates']
```

```python
launch_sites_df.loc[0,'Coordinates'] = '31°05′58″N 2°50′09″W\ufeff / \ufeff31.09951°N 2.83581°W'
```

```python
launch_sites_df.loc[0,'Coordinates']
```

```python
launch_sites_df = launch_sites_df.loc[(launch_sites_df.Coordinates != 'Mobile')]
launch_sites_df = launch_sites_df.dropna(subset=['Coordinates'])
launch_sites_df
```

```python
launch_sites_df['coordinates (dms)'] = launch_sites_df['Coordinates'].map(lambda x: str(x).split('/')[0])
launch_sites_df['lat (dms)'] = launch_sites_df['Coordinates'].map(lambda x: str(x).split('/')[0].split(' ')[0])
launch_sites_df['long (dms)'] = launch_sites_df['Coordinates'].map(lambda x: str(x).split('/')[0].split(' ')[1])
launch_sites_df['coordinates (decimal)'] = launch_sites_df['Coordinates'].map(lambda x: str(x).split('/')[1])
launch_sites_df['lat (decimal)'] = launch_sites_df['Coordinates'].map(lambda x: str(x).split('/')[1].split(' ')[1])
launch_sites_df['long (decimal)'] = launch_sites_df['Coordinates'].map(lambda x: str(x).split('/')[1].split(' ')[2])
```

```python
launch_sites_df['lat (decimal)'] = launch_sites_df['lat (decimal)'].map(lambda x: '-' + x.split('°')[0].replace('\ufeff', '') if 'S' in x else x.split('°')[0].replace('\ufeff', ''))
launch_sites_df['long (decimal)'] = launch_sites_df['long (decimal)'].map(lambda x: '-' + x.split('°')[0].replace('\ufeff', '') if 'W' in x else x.split('°')[0].replace('\ufeff', ''))
```

```python
launch_sites_df
```

```python
import math

Earth_Radius = 6378137
earth_rotational_vel = (2*math.pi*Earth_Radius)/86164
earth_rotational_vel
```

```python
launch_sites_df['rotational_vel'] = launch_sites_df['lat (decimal)'].map(lambda x: earth_rotational_vel*math.cos(math.radians(float(x))))
launch_sites_df
```
